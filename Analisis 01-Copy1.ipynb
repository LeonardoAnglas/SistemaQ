## Código general ##

```python
def procesa(archivo):

    import pandas as pd

    pd.set_option('mode.chained_assignment', None)

    df=archivo

    df1=df.iloc[:,[0,22,10,12,15,16,19,20,21]] #info principal
    df2=df.iloc[0,[3,4,5,6,7,8,9]] #encabezado
    uit=5*4950

    print("INFORMACIÓN GENERAL")
    print("-------------------")
    display(df2)

    df1.loc[df1['Estado Carta']=='En proceso',"Monto Garantizado"]=df1.loc[:,'Monto Presupuestado']
    df1.loc[df1['Estado Carta']=='Liquidación total',"Pagado"]=df1.loc[:,'Importe Factura']
    df1.loc[df1['Estado Carta']=='Liquidación parcial',"Pagado"]=df1.loc[:,'Importe Factura']

    print("")
    print("BALANCE")
    print("---------")
    print("Monto máximo (5 UIT):                S/",uit)
    montoGarantizado=df1['Monto Garantizado'].sum()
    print("Total Monto garantizado:             S/",montoGarantizado)
    totalPagado=df1['Pagado'].sum()
    print("Total Pagado:                        S/",totalPagado)
    saldo=uit-montoGarantizado-totalPagado
    print("Saldo:                               S/",saldo)
    
    print("")
    print("REEMBOLSO")
    print("---------")
    
    rta=input("¿Hay reembolso? (s/n)")

    if rta=="s"or rta=="S":
        reembolso=float(input("Ingrese el total del reembolso    S/"))
        totalPagado+=reembolso
        print("")
        print("BALANCE CORREGIDO")
        print("-----------------")
        print("Monto máximo (5 UIT):                S/",uit)
        print("Total Monto garantizado:             S/",montoGarantizado)
        print("Total Pagado:                        S/",totalPagado)
        saldo=uit-montoGarantizado-totalPagado
        print("Saldo:                               S/",saldo)

    display(df1)
    print("")
    print("OBS: Los reembolsos no están incluidos en el listado")
    print("")
    print("*************** F I N   D E L   R E P O R T E ***************")
```
